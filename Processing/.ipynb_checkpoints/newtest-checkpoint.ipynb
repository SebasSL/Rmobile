```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from skimage.measure import compare_ssim
from os import listdir, path
from tempfile import TemporaryFile

%matplotlib inline
```

```python
pklot = cv2.imread("new3.jpg",cv2.IMREAD_GRAYSCALE)

plt.imshow(pklot,cmap="gray")
plt.show()
```

```python
pklot_border = cv2.Canny(pklot,200,255)
plt.figure(figsize=(16,16))
plt.imshow(pklot_border,cmap="gray")
plt.show()
```
